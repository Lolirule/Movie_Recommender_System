```python
def extract_main_cast(obj):
    L = []
    i=0
    for x in ast.literal_eval(obj):
      if(i==3):
        break
      L.append(x['name'])
      i+=1
    return L
```

```python
## Some words like action,actions and love,loving,loved are taken as different words but as we know that they are words which mean the same
## for such words we will use a method called stem which will make the similiar meaning words the same as one of those words
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
## We try converting Textual data to vectors using vectorisation techniques.We convert them to vectors and later try to find 5 closes vectors which would be our similiar
## vectors (Similiar text)
```

```python
## We will sort the data values for each row (higest to lowest similiar),so that our first 5
```

```python
## Gettting index for a movie
new_df[new_df['title']== 'Avatar'].index[0]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances =  similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse= True,key = lambda x: x[1])[1:6]
  recommended_movies=[]
  for i in movies_list:
    recommended_movies.append([new_df.iloc[i[0]].title,i[0]])
  return recommended_movies
```
